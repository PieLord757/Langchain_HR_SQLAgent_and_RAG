```python
import numpy as np
import pandas as pd
from scipy.stats import zscore
```

```python
df = pd.read_excel('./Sample_Data_Salary_usecase.xlsx')
```

```python
print(df.head())
```

```python
df = df.rename(columns={'Unnamed: 13':'RDS_instructions'})
df.drop('Unnamed: 12',axis=1, inplace = True)
```

```python
df['DynamoDB_instructions'] = ''
#Possibly replace the table.put_item with batch.put_item for batch insertion
for i in range(len(df)):
    df.loc[i, 'DynamoDB_instructions'] = """dynamodbClient.put_item(TableName='Salary',Item={{'EMP_ID':{{'N':'{EMP_ID}'}},'work_year':{{'N':'{work_year}'}},'experience_level':{{'S':'{experience_level}'}},'employment_type':{{'S':'{employment_type}'}},'job_title':{{'S':'{job_title}'}},'salary':{{'N':'{salary}'}},'salary_currency':{{'S':'{salary_currency}'}},'salary_in_usd':{{'N':'{salary_in_usd}'}},'employee_residence':{{'S':'{employee_residence}'}},'remote_ratio':{{'N':'{remote_ratio}'}},'company_location':{{'S':'{company_location}'}},'company_size':{{'S':'{company_size}'}}}})""".format(
        EMP_ID = df.loc[i, 'EMP_ID'],
        work_year = df.loc[i, 'work_year'],
        experience_level = df.loc[i, 'experience_level'],
        employment_type = df.loc[i, 'employment_type'],
        job_title = df.loc[i, 'job_title'],
        salary = df.loc[i, 'salary'],
        salary_currency = df.loc[i, 'salary_currency'],
        salary_in_usd = df.loc[i, 'salary_in_usd'],
        employee_residence = df.loc[i, 'employee_residence'],
        remote_ratio = df.loc[i, 'remote_ratio'],
        company_location = df.loc[i, 'company_location'],
        company_size = df.loc[i, 'company_size']
        )




print(df.head())
print(df.loc[0,'DynamoDB_instructions'])
```

```python
from faker import Faker
import random

# name_df.head()

fake = Faker()
Faker.seed(42)
random.seed(42)
name_df = pd.read_excel('Person_usecase.xlsx')

target_count = 16534
existing_count = len(name_df)
remaining_count = target_count - existing_count

tech_departments = [
    "AI/ML", "Data Engineering", "Cloud", "DevOps", "QA Automation", "Security",
    "BI", "Frontend", "Backend", "Fullstack", "Java", "Python", "Data Science",
    "SRE", "Infrastructure", "Testing", "Support", "Analytics", "RPA", "Network"
]
generated = []
for i in range(remaining_count):
    name = fake.first_name()
    dept = random.choice(tech_departments)
    generated.append({"name":name, "dept":dept})
generated_df = pd.DataFrame(generated)
generated_df["emp_id"] = range(existing_count + 1, target_count + 1)

name_df["emp_id"] = range(1, existing_count+1)

# print(name_df.head())
# print(generated_df.tail())
full_df = pd.concat([name_df,generated_df],ignore_index=True)
# print(full_df.shape[0])
full_df = full_df.rename(columns={'emp_id':'EMP_ID'})
merged_df = pd.merge(full_df, df, on="EMP_ID",how="inner")
sql_template = "INSERT INTO salary VALUES ({EMP_ID}, '{name}', '{dept}', {work_year}, '{experience_level}', '{employment_type}', '{job_title}', {salary}, '{salary_currency}', {salary_in_usd}, '{employee_residence}', {remote_ratio}, '{company_location}', '{company_size}');"

merged_df['RDS_instructions'] = merged_df.apply(lambda row: sql_template.format(**row), axis=1)
print(merged_df.loc[0,"RDS_instructions"])
merged_df.to_excel("named_salary.xlsx", index=False)
```
